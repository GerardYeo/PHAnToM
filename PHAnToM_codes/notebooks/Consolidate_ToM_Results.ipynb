```python
import pandas as pd
import os
import fnmatch
import json

def find_files(directory, prefix):
    matches = []
    for root, dirnames, filenames in os.walk(directory):
        for filename in fnmatch.filter(filenames, prefix + "*.json"):
            matches.append(os.path.join(root, filename))
    return matches


def get_output(file):
    with open(file) as f:
        contents = json.load(f)

    contents['filename'] = file
    contents['source'] = 'theirs' if 'theirs' in file else 'ours'
    contents['personality'] = os.path.splitext(os.path.basename(file))[0].split('_p')[-1][1:].title()

    return pd.DataFrame.from_dict(contents, orient='index').T


output_dir = r"D:\71 PsychLLM\outs\fantom"
prefix = "REPORT_"

output = pd.DataFrame()
json_files = find_files(output_dir, prefix)
for file in json_files:
    if 'sample' in file:
        continue
    if 'archive_' in file:
        continue
    output = pd.concat([output,get_output(file)])

output
```

```python
output.to_csv(os.path.join(output_dir, 'consolidated.csv'), index=False, encoding='utf-8-sig')
```
