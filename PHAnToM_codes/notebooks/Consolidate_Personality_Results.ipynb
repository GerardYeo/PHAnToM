```python
import pandas as pd
import os
import fnmatch
from ast import literal_eval


def find_files(directory, prefix):
    matches = []
    for root, dirnames, filenames in os.walk(directory):
        for filename in fnmatch.filter(filenames, prefix + "*.json"):
            matches.append(os.path.join(root, filename))
    return matches


def get_output(file):
    with open(file) as f:
        lines = f.readlines()
    
    row_output = {
        'filename':file,
        'model': os.path.basename(file).split('_')[1],
        'personality': os.path.splitext(os.path.basename(file))[0].split('_p')[-1][1:].title()
    }

    brute = literal_eval(lines[0].split('mean:')[-1].split('std:')[0].split('[')[-1].split(']')[0])
    for personality,score in brute:
        row_output[f'{personality}_mean']=score

    brute = literal_eval(lines[0].split('std:')[-1].split('[')[-1].split(']')[0])
    for personality,score in brute:
        row_output[f'{personality}_std']=score

    return pd.DataFrame.from_dict(row_output, orient='index').T


output_dir = r"D:\71 PsychLLM\outs\personality"
prefix = "REPORT_"

output = pd.DataFrame()
json_files = find_files(output_dir, prefix)
for file in json_files:
    output = pd.concat([output,get_output(file)])

output
```

```python
output.to_csv(os.path.join(output_dir, 'consolidated.csv'), index=False, encoding='utf-8-sig')    
```
