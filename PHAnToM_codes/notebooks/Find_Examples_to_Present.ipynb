```python
"""
Finding examples to display for qualitative analysis.

Criteria:
1. Y/N task
2. Answer is No (Hypothesis: attention needed for whole conversation)
3. Some personality-v right, some wrong.
"""

import pandas as pd
import os
import fnmatch
import json

def find_files(directory, prefix):
    matches = []
    for root, dirnames, filenames in os.walk(directory):
        for filename in fnmatch.filter(filenames, prefix + "*.json"):
            matches.append(os.path.join(root, filename))
    return matches


focus_tasks = ['tom:info_accessibility:binary', 'tom:answerability:binary']
def get_output(file):

    with open(file) as f:
        contents = json.load(f)

    model = contents['model']
    personality = os.path.splitext(os.path.basename(file))[0].split('_p')[-1][1:].title()
    
    model_output = []
    for item in contents['results']:
        if (item['question_type'] in focus_tasks) and ('no' in item['correct_answer']):
            model_output.append([
                item['question'],
                item['question_type'],
                item['correct_answer'],
                item['set_id'],
                item['result']
            ])
    
    model_output = pd.DataFrame(
        model_output, 
        columns=['question','question_type','correct_answer','set_id',f'{model}_{personality}']
    )
    
    return model_output


output_dir = r"D:\71 PsychLLM\outs\fantom"
prefix = "evaluated_responses_"

counter = 0
output = pd.DataFrame()
json_files = find_files(output_dir, prefix)
for file in json_files:
    if 'p2_theirs' in file:
        continue
    if 'sample' in file:
        continue
    if 'mistral' not in file:
        continue
    output = pd.concat([output,get_output(file)],axis=1)
    counter+=1

output
```

```python
final = pd.concat([output.iloc[:,:4], output[['mistral-instruct_Agreeableness',
 'mistral-instruct_Conscientiousness',
 'mistral-instruct_Extraversion',
 'mistral-instruct_Machiavellianism',
 'mistral-instruct_Narcissism',
 'mistral-instruct_Neuroticism',
 'mistral-instruct_None',
 'mistral-instruct_Openness',
 'mistral-instruct_Psychopathy',
 'mistral-instruct_Task-Specific']]], axis=1)
final
```

```python
final.to_csv(os.path.join(output_dir,'no_examples_mistral.csv'))
```
